# 🛒 Association Rules: Visualization Lab

Welcome to the Apriori and Association Rules Lab. We will mine frequent itemsets and visualize rule networks.

## 1. Theory Overview

We use the Apriori algorithm to find Frequent Itemsets, and then generate rules (X -> Y) based on Confidence and Lift.

## 2. Visualization Playground
Let's create a synthetic grocery dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Synthetic one-hot encoded dataset
data = {
    'Milk': [1, 1, 0, 1, 0, 1, 1],
    'Bread': [1, 1, 1, 1, 1, 0, 1],
    'Butter': [1, 0, 1, 1, 1, 0, 0],
    'Eggs': [0, 1, 0, 1, 0, 1, 1],
    'Beer': [0, 0, 1, 0, 1, 0, 0],
    'Diapers': [0, 0, 1, 0, 1, 0, 0]
}
df = pd.DataFrame(data)
print("Transaction Data:")
print(df)
```

## 3. Scikit-Learn Implementation
Scikit-Learn doesn't support Apriori, so we use `mlxtend`.

```python
# 1. Find Frequent Itemsets
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)

# 2. Generate Rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Clean up the names for visualization
rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x)))

print("Association Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

## 4. Hyperparameter Impact Study
Visualizing Lift Matrix via Heatmap.

```python
pivot = rules.pivot(index='antecedents', columns='consequents', values='lift')

plt.figure(figsize=(8, 6))
sns.heatmap(pivot, annot=True, cmap="YlOrRd", fmt=".2f")
plt.title("Lift Matrix Heatmap")
plt.show()
```
