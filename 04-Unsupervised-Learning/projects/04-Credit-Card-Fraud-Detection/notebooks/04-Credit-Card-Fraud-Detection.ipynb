# 💳 Credit Card Fraud Detection

Using Isolation Forest to detect anomalies (frauds) in credit card transactions.

```
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt

# Normal transactions
X, _ = make_blobs(n_samples=500, centers=1, cluster_std=0.5, random_state=42)
# Fraud transactions (outliers)
outliers = np.random.uniform(low=-4, high=4, size=(20, 2))
data = np.vstack([X, outliers])

iso_forest = IsolationForest(contamination=0.04, random_state=42)
preds = iso_forest.fit_predict(data)

plt.scatter(data[:, 0], data[:, 1], c=preds, cmap='coolwarm')
plt.title('Anomaly Detection (Red = Fraud)')
plt.show()
```
