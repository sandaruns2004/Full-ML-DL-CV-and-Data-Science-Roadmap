# 04 - Regression Metrics

In this notebook, we explore how to mathematically evaluate continuous predictions using MAE, MSE, RMSE, and R-Squared.

## Concept Overview
Regression metrics measure *distance*. How far off was the prediction from the truth? Different metrics punish that distance in different ways.

## Scikit-Learn Implementation
Let's calculate the core metrics on a dummy dataset.

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_true = np.array([100, 150, 200, 250, 300])
y_pred = np.array([110, 140, 210, 240, 450]) # Notice the massive outlier on the last prediction

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

print(f'MAE:  {mae:.2f}')
print(f'MSE:  {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'R2:   {r2:.4f}')
```

Notice how the MAE is 38, but the RMSE is 67. The RMSE is heavily pulled upward by that single massive mistake (predicted 450, actual 300).

## Visual Demonstration: Residual Plots
A metric is just a single number. To truly understand a regression model, you must plot its residuals (errors).

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Generate some semi-realistic data
np.random.seed(42)
y_true_large = np.random.uniform(50, 500, 200)
# Model systematically underpredicts high values (non-linear error)
y_pred_large = y_true_large * 0.9 + np.random.normal(0, 15, 200)

residuals = y_true_large - y_pred_large

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1. Actual vs Predicted
axes[0].scatter(y_true_large, y_pred_large, alpha=0.6)
axes[0].plot([50, 500], [50, 500], color='red', linestyle='--') # Perfect prediction line
axes[0].set_title('Actual vs Predicted')
axes[0].set_xlabel('Actual Values')
axes[0].set_ylabel('Predicted Values')

# 2. Residual Plot
axes[1].scatter(y_pred_large, residuals, alpha=0.6)
axes[1].axhline(y=0, color='red', linestyle='--')
axes[1].set_title('Residual Plot')
axes[1].set_xlabel('Predicted Values')
axes[1].set_ylabel('Residual (Error)')

plt.show()
```

Looking at the Residual Plot, we can clearly see a pattern. As the predicted values get higher, the errors become increasingly positive. This means our model is fundamentally missing a trend in the data, despite what the R-squared score might say!

## Industry Notes
Never report MSE to a business stakeholder. It is measured in 'squared units' (e.g., squared dollars). Always report MAE or RMSE, which are in the original units.
