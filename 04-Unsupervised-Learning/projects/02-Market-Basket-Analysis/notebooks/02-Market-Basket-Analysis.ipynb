# 🛒 Market Basket Analysis

Instead of mlxtend, we'll manually demonstrate the concept of item co-occurrence (which underlies Apriori).

```
import pandas as pd
from itertools import combinations
from collections import Counter

transactions = [
    ['Milk', 'Bread', 'Eggs'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread', 'Butter', 'Eggs'],
    ['Milk', 'Eggs']
]

pair_counts = Counter()
for t in transactions:
    pair_counts.update(combinations(sorted(t), 2))

print("Most common item pairs:")
for pair, count in pair_counts.most_common():
    print(f"{pair}: {count}")
```
