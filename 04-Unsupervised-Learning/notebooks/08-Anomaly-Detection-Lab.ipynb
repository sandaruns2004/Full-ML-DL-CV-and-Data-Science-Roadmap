## 2. Visualization Playground (1D Stats)
Let's visualize IQR using a Box Plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 3)

data = np.random.normal(50, 10, 200)
outliers = np.array([100, 110, -10])
data_with_outliers = np.concatenate([data, outliers])

sns.boxplot(x=data_with_outliers, color="skyblue")
plt.title("Box Plot highlighting Statistical Anomalies (IQR Method)")
plt.show()
```

## 3. Scikit-Learn Implementation: Isolation Forest vs LOF
Let's generate a dataset with a dense cluster, a sparse cluster, and some global anomalies.

```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

plt.rcParams['figure.figsize'] = (15, 6)

# 1. Dense Cluster
dense = np.random.normal(0, 1, (200, 2))
# 2. Sparse Cluster
sparse = np.random.normal(10, 4, (50, 2))
# 3. Anomalies
outliers_global = np.random.uniform(low=-10, high=20, size=(10, 2))

X = np.vstack([dense, sparse, outliers_global])

fig, axes = plt.subplots(1, 2)

# --- Isolation Forest ---
iso = IsolationForest(contamination=0.05, random_state=42)
iso_labels = iso.fit_predict(X)
axes[0].scatter(X[:, 0], X[:, 1], c=iso_labels, cmap='coolwarm', s=30, edgecolor='k')
axes[0].set_title("Isolation Forest (Global Anomalies)")

# --- Local Outlier Factor ---
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
lof_labels = lof.fit_predict(X)
axes[1].scatter(X[:, 0], X[:, 1], c=lof_labels, cmap='coolwarm', s=30, edgecolor='k')
axes[1].set_title("Local Outlier Factor (Contextual Anomalies)")

plt.show()
```

## 4. Hyperparameter Impact Study
Visualizing LOF density circles (Radius proportional to LOF score).

```python
plt.rcParams['figure.figsize'] = (10, 8)

lof_scores = -lof.negative_outlier_factor_
radius = (lof_scores - lof_scores.min()) / (lof_scores.max() - lof_scores.min())

plt.scatter(X[:, 0], X[:, 1], color='k', s=10, label='Data points')
plt.scatter(X[:, 0], X[:, 1], s=2000 * radius, edgecolors='r', facecolors='none', label='LOF Score Radius')
plt.title("LOF Scores Visualized as Bubble Radii")
plt.legend()
plt.show()
```
