# Project 5: Imbalanced Data Challenge\n\n**Objective**: Catch fraud in a highly imbalanced dataset using Algorithmic (Class Weights) and Data (SMOTE) interventions.

## Setup & Data Generation

```python
import numpy as np\nfrom sklearn.datasets import make_classification\nfrom sklearn.model_selection import train_test_split\nfrom sklearn.ensemble import RandomForestClassifier\nfrom sklearn.metrics import classification_report, f1_score\n\n# 98% Normal, 2% Fraud\nX, y = make_classification(n_samples=5000, weights=[0.98, 0.02], random_state=42)\nX_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

## Intervention 1: Class Weights

```python
weighted = RandomForestClassifier(class_weight='balanced', random_state=42)\nweighted.fit(X_train, y_train)\npreds_weight = weighted.predict(X_test)\nprint("--- Class Weighted Model ---")\nprint(f"F1 Score (Fraud Class): {f1_score(y_test, preds_weight):.4f}")
```

## Intervention 2: SMOTE Pipeline\n*Crucial: We must use the imblearn Pipeline, not the sklearn Pipeline, to ensure SMOTE is only applied during fit, not during predict.*

```python
from imblearn.over_sampling import SMOTE\nfrom imblearn.pipeline import Pipeline\n\nsmote_pipeline = Pipeline([\n    ('smote', SMOTE(random_state=42)),\n    ('rf', RandomForestClassifier(random_state=42))\n])\n\nsmote_pipeline.fit(X_train, y_train)\npreds_smote = smote_pipeline.predict(X_test)\nprint("--- SMOTE Model ---")\nprint(f"F1 Score (Fraud Class): {f1_score(y_test, preds_smote):.4f}")
```

## Conclusion

```python
print("Final Report (SMOTE Model):")\nprint(classification_report(y_test, preds_smote, target_names=['Normal', 'Fraud']))
```
