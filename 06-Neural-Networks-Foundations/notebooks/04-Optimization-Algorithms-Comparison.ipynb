# 04 - Optimizers Comparison
Compare SGD, Momentum, RMSProp, and Adam visually on a dummy loss surface.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Visualize convergence speeds hypothetically
epochs = np.arange(100)
sgd_loss = np.exp(-0.02 * epochs) + 0.1 * np.random.randn(100)
momentum_loss = np.exp(-0.05 * epochs) + 0.05 * np.random.randn(100)
adam_loss = np.exp(-0.1 * epochs) + 0.01 * np.random.randn(100)

plt.figure(figsize=(10, 6))
plt.plot(epochs, sgd_loss, label='SGD', alpha=0.7)
plt.plot(epochs, momentum_loss, label='Momentum', alpha=0.7)
plt.plot(epochs, adam_loss, label='Adam', linewidth=2)
plt.title('Optimizer Convergence Comparison')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
