# 🖼️ Image Compression with PCA

Using Principal Component Analysis to reduce the dimensionality of an image, effectively compressing it.

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_sample_image

img = load_sample_image('china.jpg')
gray_img = np.mean(img, axis=2)

pca = PCA(n_components=50)
compressed_img = pca.fit_transform(gray_img)
reconstructed_img = pca.inverse_transform(compressed_img)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(gray_img, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(reconstructed_img, cmap='gray')
ax[1].set_title('Compressed (50 PCs)')
plt.show()
```
