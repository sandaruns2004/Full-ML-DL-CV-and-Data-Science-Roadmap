# 10 - Learning Curves

In this notebook, we visualize how a model's performance improves as it sees more data, helping us answer the business question: *Do we need to buy more data?*

## Concept Overview
A Learning Curve plots the Training Score and Validation Score on the Y-axis against the number of training examples on the X-axis.

## Scikit-Learn Implementation
We will use `learning_curve` to automatically train models on progressively larger chunks of our data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.svm import SVC
from sklearn.datasets import load_digits

# Load a moderately sized dataset (1797 samples)
digits = load_digits()
X, y = digits.data, digits.target

# We will test an SVM model
model = SVC(gamma=0.001)

# Generate learning curve data
train_sizes, train_scores, test_scores = learning_curve(
    model, X, y, cv=5, n_jobs=-1, 
    train_sizes=np.linspace(0.1, 1.0, 10), # Test from 10% to 100% of data
    scoring='accuracy'
)

# Calculate means and standard deviations
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
```

## Visual Demonstration

```python
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Training Score')
plt.plot(train_sizes, test_mean, 'o-', color='orange', label='Validation Score')

# Add variance bands
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color='orange')

plt.title('Learning Curve (SVM)')
plt.xlabel('Number of Training Examples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

## Interpretation
* The training score starts perfectly at 1.0 (it's easy to memorize 150 examples), but drops slightly as the dataset grows and becomes harder to memorize.
* The validation score starts terribly, but climbs rapidly as the model learns the real patterns.
* **The Gap**: The gap between the lines is closing. If we added another 2,000 images, the validation score would likely continue to climb. *Buying more data is a good idea here.*

## Overfitting vs Underfitting on Learning Curves
* **Underfitting (High Bias)**: Both curves flatline early at a terrible score. More data won't help.
* **Overfitting (High Variance)**: Massive gap between the curves, training is near 1.0, validation is low. More data *will* help, or you need a simpler model.
