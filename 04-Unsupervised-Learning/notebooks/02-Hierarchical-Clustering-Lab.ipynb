## 1. Theory Overview

Hierarchical clustering builds a tree of clusters (dendrogram). It doesn't require pre-specifying $K$.

## 2. Visualization Playground
Let's build a dendrogram using SciPy.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import make_blobs

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Generate simple 2D data
X, y = make_blobs(n_samples=50, centers=3, cluster_std=1.0, random_state=42)

# Calculate the linkage matrix using Ward's method
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.title("Hierarchical Clustering Dendrogram (Ward Linkage)")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.)
plt.axhline(y=10, color='r', linestyle='--', label='Cut Threshold for K=3')
plt.legend()
plt.show()
```

## 5. Scikit-Learn Implementation
Using `AgglomerativeClustering`.

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
labels = hc.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set2', s=100, edgecolor='k')
plt.title("Agglomerative Clustering (K=3)")
plt.show()
```

## 6. Hyperparameter Impact Study
Comparing Linkage Methods: Single vs Complete vs Ward.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
methods = ['single', 'complete', 'ward']

for i, method in enumerate(methods):
    model = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage=method)
    labels = model.fit_predict(X)
    axes[i].scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10', s=50, edgecolor='k')
    axes[i].set_title(f"{method.capitalize()} Linkage")

plt.tight_layout()
plt.show()
```

## 7. Real Dataset Case Study
Using the Wine dataset.

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler

wine = load_wine()
X_wine = StandardScaler().fit_transform(wine.data)

Z_wine = linkage(X_wine, method='complete')
plt.figure(figsize=(10, 5))
plt.title("Wine Dataset Dendrogram (Complete Linkage)")
dendrogram(Z_wine, truncate_mode='lastp', p=12)
plt.show()
```
