# 🎯 K-Means Clustering: Visualization Lab

Welcome to the K-Means Visualization Laboratory. In this notebook, we will explore the internal mechanics of K-Means, visualize the Elbow Method, and see how centroids converge.

## 1. Theory Overview

K-Means aims to partition $N$ observations into $K$ clusters in which each observation belongs to the cluster with the nearest mean. It minimizes the Within-Cluster Sum of Squares (WCSS).

## 2. Visualization Playground

Let's generate some synthetic data and visualize how K-Means iteratively moves centroids.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Set aesthetic style
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (10, 6)

# Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.8, random_state=42)

plt.scatter(X[:, 0], X[:, 1], s=30, alpha=0.7)
plt.title("Synthetic Data for Clustering")
plt.show()
```

### Centroid Convergence Animation Logic
Here we manually plot a few steps of the algorithm to simulate the animation.

```python
# Step-by-step K-Means to visualize centroid movement
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, max_iter in enumerate([1, 2, 5]):
    kmeans = KMeans(n_clusters=4, init='random', max_iter=max_iter, n_init=1, random_state=42)
    kmeans.fit(X)
    
    axes[i].scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', s=30, alpha=0.6)
    axes[i].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
                    s=200, c='red', marker='X', edgecolor='black')
    axes[i].set_title(f"Iteration {max_iter}")

plt.tight_layout()
plt.show()
```

## 3. From Scratch Implementation

```python
# A simplified K-Means from scratch
def kmeans_scratch(X, k=4, max_iters=100):
    np.random.seed(42)
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]
    
    for _ in range(max_iters):
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])
        
        if np.allclose(centroids, new_centroids):
            break
        centroids = new_centroids
        
    return centroids, labels

scratch_centroids, scratch_labels = kmeans_scratch(X, k=4)
print("Scratch implementation converged successfully!")
```

## 4. NumPy Implementation
Vectorized operations make it fast. The `distances` computation above uses NumPy broadcasting trick: `X[:, np.newaxis] - centroids`.

## 5. Scikit-Learn Implementation
The industry standard.

```python
from sklearn.cluster import KMeans

kmeans_sk = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=300, random_state=42)
labels_sk = kmeans_sk.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels_sk, cmap='viridis', s=30, alpha=0.6)
plt.scatter(kmeans_sk.cluster_centers_[:, 0], kmeans_sk.cluster_centers_[:, 1], s=200, c='red', marker='X', edgecolor='k')
plt.title("Scikit-Learn K-Means++")
plt.show()
```

## 6. Hyperparameter Impact Study
Let's study the Elbow Method to find the optimal $K$.

```python
inertias = []
K_range = range(1, 10)

for k in K_range:
    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    model.fit(X)
    inertias.append(model.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, marker='o', linestyle='--', color='b')
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Inertia)")
plt.axvline(x=4, color='r', linestyle=':', label='Optimal K (Elbow)')
plt.legend()
plt.show()
```

## 7. Real Dataset Case Study
Using the Iris dataset.

```python
from sklearn.datasets import load_iris

iris = load_iris()
X_iris = iris.data[:, 2:] # Use Petal Length and Width for 2D visual

kmeans_iris = KMeans(n_clusters=3, n_init=10, random_state=42)
iris_labels = kmeans_iris.fit_predict(X_iris)

plt.scatter(X_iris[:, 0], X_iris[:, 1], c=iris_labels, cmap='Set1', s=50, edgecolor='k')
plt.title("K-Means on Iris Dataset (Petal Features)")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```
