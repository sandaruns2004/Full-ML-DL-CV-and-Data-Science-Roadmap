# Project 6: End-To-End Model Deployment (Drift Monitoring)\n\n**Objective**: Simulate an ML model in production over 6 months and use Evidently AI to automatically generate a Data Drift report when a feature breaks.

## Setup & Installation\nEnsure `evidently` is installed.\n`pip install evidently`

```python
import numpy as np\nimport pandas as pd\nfrom evidently.report import Report\nfrom evidently.metric_preset import DataDriftPreset
```

## Simulating Production Data

```python
np.random.seed(42)\n\n# 1. Training Data (Reference)\ntraining_data = pd.DataFrame({\n    'age': np.random.normal(40, 10, 1000),\n    'income': np.random.normal(60000, 15000, 1000),\n    'credit_score': np.random.normal(700, 50, 1000)\n})\n\n# 2. Production Data (Month 6) - The 'income' feature drifts significantly\nproduction_data = pd.DataFrame({\n    'age': np.random.normal(40, 10, 1000),\n    'income': np.random.normal(90000, 20000, 1000), # Huge drift due to inflation/new user base\n    'credit_score': np.random.normal(695, 50, 1000)\n})
```

## Generating the Drift Report with Evidently AI

```python
print("Generating Evidently Data Drift Report...")\ndrift_report = Report(metrics=[DataDriftPreset()])\ndrift_report.run(reference_data=training_data, current_data=production_data)\n\n# Save to an HTML file\nhtml_path = 'production_drift_report.html'\ndrift_report.save_html(html_path)\nprint(f"Report saved to {html_path}. Open this file in your web browser!")
```

## Review\nWhen you open the HTML report, you will see a massive visual breakdown proving that the `income` column has drifted, while `age` and `credit_score` remained stable.
