# 🏆 Sequence Models Benchmark

An advanced notebook benchmarking RNNs, LSTMs, GRUs, and 1D CNNs against each other for parameter efficiency, memory footprint, and training speed.

```
import torch
import time
import torch.nn as nn

def benchmark_model(model, input_tensor):
    model.eval()
    # Warmup
    for _ in range(5):
        _ = model(input_tensor)
        
    # Timing
    start = time.time()
    for _ in range(50):
        _ = model(input_tensor)
    end = time.time()
    
    params = sum(p.numel() for p in model.parameters())
    print(f"Model: {model.__class__.__name__}")
    print(f"Parameters: {params:,}")
    print(f"Time (50 passes): {(end-start)*1000:.2f} ms\n")

# Compare simple layers
seq_len = 1000
batch = 16
dim = 64

x = torch.randn(batch, seq_len, dim)

benchmark_model(nn.RNN(dim, dim, batch_first=True), x)
benchmark_model(nn.LSTM(dim, dim, batch_first=True), x)
benchmark_model(nn.GRU(dim, dim, batch_first=True), x)
benchmark_model(nn.Conv1d(dim, dim, kernel_size=3, padding=1), x.transpose(1, 2))
```
