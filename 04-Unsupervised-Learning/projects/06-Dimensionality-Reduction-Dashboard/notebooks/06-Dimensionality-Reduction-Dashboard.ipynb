# 🎛️ Dimensionality Reduction Dashboard

Comparing PCA and t-SNE on high-dimensional data (digits dataset).

```
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

digits = load_digits()
X, y = digits.data, digits.target

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='jet', alpha=0.6)
ax[0].set_title('PCA')
ax[1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='jet', alpha=0.6)
ax[1].set_title('t-SNE')
plt.show()
```
