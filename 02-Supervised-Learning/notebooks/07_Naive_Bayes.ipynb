# 📩 Naive Bayes Workshop
Naive Bayes is highly scalable, fast, and relies on Bayes' Theorem. It's fantastic for text classification.

## 1. Problem Overview
We will classify text messages as Spam or Ham.

## 2. Dataset Exploration

```python
import pandas as pd

texts = ['free money now', 'hello mom', 'win a free car', 'meeting at 5', 'claim your prize']
labels = [1, 0, 1, 0, 1] # 1 is Spam
df = pd.DataFrame({'text': texts, 'label': labels})
display(df)
```

## 3. Data Preprocessing
Converting text to Bag-of-Words.

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text']).toarray()
print(vectorizer.get_feature_names_out())
print(X)
```

## 6. Scikit-Learn Workflow

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X, df['label'])

test_text = ['free car']
test_vec = vectorizer.transform(test_text)
print('Prediction for "free car":', 'Spam' if model.predict(test_vec)[0] == 1 else 'Ham')
```

## 9. Failure Cases
"Naive" means it assumes words are completely independent. It fails to understand word order or phrases (e.g., "not bad" might be parsed as negative because "bad" is usually negative).
