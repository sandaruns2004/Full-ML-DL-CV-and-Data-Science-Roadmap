# 05 - Vanishing & Exploding Gradients
Demonstrate how gradients vanish when using Sigmoid in deep networks.

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def sigmoid_derivative(z):
    s = 1 / (1 + np.exp(-z))
    return s * (1 - s)

# A deep network with 100 layers
gradient = 1.0
gradients = []

for i in range(100):
    # Assume input z is drawn from normal dist
    z = np.random.randn()
    # Multiply local gradient (chain rule)
    local_grad = sigmoid_derivative(z)
    gradient *= local_grad
    gradients.append(gradient)

plt.figure(figsize=(10, 5))
plt.plot(gradients, color='red')
plt.title('Vanishing Gradient over 100 Layers using Sigmoid')
plt.xlabel('Layer Depth (Backward)')
plt.ylabel('Gradient Magnitude')
plt.yscale('log')
plt.grid(True)
plt.show()
```
