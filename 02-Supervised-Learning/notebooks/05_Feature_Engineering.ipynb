# 🛠️ Feature Engineering

Feature engineering involves modifying your dataset to improve model performance, such as scaling features or encoding categorical variables.

```
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

data = pd.DataFrame({
    'Age': [25, 30, 45, 22],
    'City': ['NY', 'LA', 'NY', 'SF'],
    'Income': [50000, 80000, 120000, 40000]
})

ct = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Age', 'Income']),
        ('cat', OneHotEncoder(), ['City'])
    ]
)

transformed = ct.fit_transform(data)
print(transformed)
```
