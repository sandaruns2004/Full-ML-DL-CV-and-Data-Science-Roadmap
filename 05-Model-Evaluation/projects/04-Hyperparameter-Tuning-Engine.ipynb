# Project 4: Hyperparameter Tuning Engine\n\n**Objective**: Compare Grid Search vs Random Search on a Random Forest model. Prove that Random Search finds excellent parameters significantly faster.

## Setup & Imports

```python
import numpy as np\nimport time\nfrom sklearn.datasets import fetch_california_housing\nfrom sklearn.ensemble import RandomForestRegressor\nfrom sklearn.model_selection import GridSearchCV, RandomizedSearchCV\nfrom scipy.stats import randint
```

## Loading Data

```python
# Using a smaller subset of California Housing for speed\nX, y = fetch_california_housing(return_X_y=True)\nX, y = X[:2000], y[:2000]\nmodel = RandomForestRegressor(random_state=42)
```

## 1. The Exhaustive Grid Search

```python
param_grid = {\n    'n_estimators': [50, 100, 150],\n    'max_depth': [5, 10, 15, None],\n    'min_samples_split': [2, 5, 10]\n}\n# 3 x 4 x 3 = 36 combinations * 3 Folds = 108 model fits\n\nstart_time = time.time()\ngrid = GridSearchCV(model, param_grid, cv=3, n_jobs=-1)\ngrid.fit(X, y)\ngrid_time = time.time() - start_time\n\nprint(f"Grid Search Time:  {grid_time:.2f} seconds")\nprint(f"Grid Search Score (R2): {grid.best_score_:.4f}")
```

## 2. The Efficient Random Search

```python
param_dist = {\n    'n_estimators': randint(50, 150),\n    'max_depth': randint(5, 20),\n    'min_samples_split': randint(2, 12)\n}\n# Try only 10 combinations * 3 Folds = 30 model fits\n\nstart_time = time.time()\nrandom_search = RandomizedSearchCV(model, param_distributions=param_dist, n_iter=10, cv=3, random_state=42, n_jobs=-1)\nrandom_search.fit(X, y)\nrandom_time = time.time() - start_time\n\nprint(f"Random Search Time:  {random_time:.2f} seconds")\nprint(f"Random Search Score (R2): {random_search.best_score_:.4f}")
```

## Conclusion

```python
print(f"Speedup: Random Search was {grid_time / random_time:.1f}x faster.")\nprint(f"Performance Difference: {abs(grid.best_score_ - random_search.best_score_):.4f} R2 points.")\nprint("Random Search achieves near-identical performance in a fraction of the time.")
```
