# 🐍 Python for Data Science

Python is the lingua franca of Data Science. In this notebook, we'll quickly review some of the most critical Python features used heavily in data manipulation.

## 1. Dictionaries and Lists
Data often comes in JSON format, which maps perfectly to Python lists of dictionaries.

```python
users = [
    {"id": 1, "name": "Alice", "role": "Admin", "active": True},
    {"id": 2, "name": "Bob", "role": "User", "active": False},
    {"id": 3, "name": "Charlie", "role": "User", "active": True}
]

# Extract just the names of active users
active_users = []
for user in users:
    if user["active"]:
        active_users.append(user["name"])
        
print("Active users:", active_users)
```

## 2. List Comprehensions
We can rewrite the above loop in a much cleaner, 'Pythonic' way.

```python
active_users_comp = [user["name"] for user in users if user["active"]]
print("Active users (via comprehension):", active_users_comp)
```

## 3. Lambda Functions
Anonymous, single-line functions are very useful when applying transformations to data frames.

```python
prices = [10.99, 25.50, 5.00, 100.00]
# Apply a 10% discount to all prices
discounted = list(map(lambda x: x * 0.9, prices))
print("Discounted Prices:", [round(p, 2) for p in discounted])
```
