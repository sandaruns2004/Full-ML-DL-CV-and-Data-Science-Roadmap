# 01 - Activation Functions Visualizer
In this notebook, we visualize the most common activation functions and their derivatives.

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_activation(name, func, deriv):
    z = np.linspace(-5, 5, 100)
    y = func(z)
    dy = deriv(z)
    
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(z, y, label=name, color='blue', linewidth=2)
    plt.title(f'{name} Function')
    plt.grid(True)
    
    plt.subplot(1, 2, 2)
    plt.plot(z, dy, label='Derivative', color='red', linewidth=2, linestyle='--')
    plt.title(f'{name} Derivative')
    plt.grid(True)
    plt.show()
```

### 1. Sigmoid

```
def sigmoid(z): return 1 / (1 + np.exp(-z))
def sigmoid_deriv(z): return sigmoid(z) * (1 - sigmoid(z))
plot_activation('Sigmoid', sigmoid, sigmoid_deriv)
```

### 2. ReLU

```
def relu(z): return np.maximum(0, z)
def relu_deriv(z): return np.where(z > 0, 1, 0)
plot_activation('ReLU', relu, relu_deriv)
```

### 3. Tanh

```
def tanh(z): return np.tanh(z)
def tanh_deriv(z): return 1 - np.tanh(z)**2
plot_activation('Tanh', tanh, tanh_deriv)
```
