# Loan Approval Classification using Bagging

## Business Problem
Predict credit default risk using Bagging to ensure stable decision boundaries and avoid overfitting to specific credit histories.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, confusion_matrix, classification_report
```

## Dataset Generation (Mock Credit Data)

```python
# Generate imbalanced dataset mimicking default rates (10% default)
X, y = make_classification(n_samples=10000, n_features=20, n_informative=12, 
                           weights=[0.9, 0.1], random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

## Model Training (Handling Class Imbalance)

```python
# We use class_weight='balanced' in the base estimator to handle the 90/10 split
base_clf = DecisionTreeClassifier(max_depth=None, class_weight='balanced')
bagging_clf = BaggingClassifier(estimator=base_clf, n_estimators=200, oob_score=True, random_state=42, n_jobs=-1)

bagging_clf.fit(X_train, y_train)
print(f"OOB Score: {bagging_clf.oob_score_:.4f}")
```

## Evaluation & Risk Analysis

```python
y_pred = bagging_clf.predict(X_test)
y_prob = bagging_clf.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
print(f"ROC AUC Score: {roc_auc_score(y_test, y_prob):.4f}")

import seaborn as sns
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix: Default Prediction')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
