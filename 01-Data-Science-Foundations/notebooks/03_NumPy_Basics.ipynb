# 🔢 NumPy Basics

NumPy (Numerical Python) is the foundational library for scientific computing in Python. It provides high-performance multidimensional array objects.

## 1. Creating Arrays
NumPy arrays are much faster and more memory-efficient than standard Python lists.

```python
import numpy as np

arr_1d = np.array([1, 2, 3, 4, 5])
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
zeros = np.zeros((3, 3))
range_arr = np.arange(0, 10, 2)

print("1D Array:\n", arr_1d)
print("\n2D Array Shape:", arr_2d.shape)
print("\nZeros:\n", zeros)
print("\nRange Array:", range_arr)
```

## 2. Vectorized Operations
You can perform mathematical operations directly on the entire array without writing loops.

```python
prices = np.array([100, 200, 300])
taxes = prices * 0.15
total = prices + taxes

print("Base Prices:", prices)
print("Taxes:", taxes)
print("Total Cost:", total)
```

## 3. Slicing and Filtering
We can easily slice arrays or filter them using boolean conditions.

```python
data = np.random.randint(1, 100, size=15)
print("Full Data:", data)

# Boolean Indexing
high_values = data[data > 80]
print("Values > 80:", high_values)
```
