# 📊 EDA and Visualization

Exploratory Data Analysis (EDA) is how we get to know our data. Visualizations are the sharpest tool in the EDA toolkit. We use `matplotlib` for foundational plotting and `seaborn` for beautiful statistical graphs.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set seaborn style
sns.set_theme(style="darkgrid")
```

## 1. Scatter Plots & Correlation
Scatter plots help us identify relationships between two continuous variables.

```python
# Synthetic Data
np.random.seed(0)
study_hours = np.random.normal(10, 3, 100)
exam_scores = (study_hours * 5) + np.random.normal(0, 10, 100)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=study_hours, y=exam_scores, color='purple', alpha=0.7)
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.show()
```

## 2. Histograms & Distributions
Histograms show how data is distributed.

```python
plt.figure(figsize=(8, 5))
sns.histplot(exam_scores, bins=15, kde=True, color='teal')
plt.title('Distribution of Exam Scores')
plt.xlabel('Score')
plt.show()
```

## 3. Boxplots
Boxplots are excellent for spotting outliers and comparing distributions across categories.

```python
departments = ['HR'] * 30 + ['Engineering'] * 50 + ['Sales'] * 40
salaries = np.concatenate([
    np.random.normal(60000, 5000, 30), 
    np.random.normal(90000, 15000, 50), 
    np.random.normal(70000, 10000, 40)
])

df = pd.DataFrame({'Department': departments, 'Salary': salaries})

plt.figure(figsize=(8, 5))
sns.boxplot(x='Department', y='Salary', data=df, palette='Set2')
plt.title('Salary Distribution by Department')
plt.show()
```
