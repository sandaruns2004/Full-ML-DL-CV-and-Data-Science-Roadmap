# 📈 Linear and Polynomial Regression

Regression models are used to predict continuous outcomes. Linear regression fits a straight line, while Polynomial regression fits a curve.

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25])

# Linear
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Polynomial
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)

plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), label='Linear')
plt.plot(X, poly_reg.predict(X_poly), label='Polynomial', color='green')
plt.legend()
plt.show()
```
