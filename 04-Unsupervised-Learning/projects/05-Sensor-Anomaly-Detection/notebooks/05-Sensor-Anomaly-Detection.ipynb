# 🏭 Sensor Anomaly Detection

Detecting failing industrial sensors using Local Outlier Factor (LOF).

```
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_circles
import numpy as np
import matplotlib.pyplot as plt

X, _ = make_circles(n_samples=300, noise=0.05, factor=0.5, random_state=42)
outliers = np.random.uniform(low=-1.5, high=1.5, size=(15, 2))
X = np.vstack([X, outliers])

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
preds = lof.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=preds, cmap='Paired')
plt.title('Sensor Anomalies with LOF')
plt.show()
```
