# 🌌 t-SNE & UMAP: Visualization Lab

Welcome to the non-linear dimensionality reduction lab! We will compare PCA, t-SNE, and UMAP on the Digits dataset.

## 1. Theory Overview

PCA is linear and preserves global variance. t-SNE is non-linear and preserves local neighborhoods but is slow. UMAP is non-linear, preserves both local and global structure, and is extremely fast.

## 2. Visualization Playground
We will load the digits dataset (8x8 images = 64 dimensions).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
import time

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (18, 6)

digits = load_digits()
X = digits.data
y = digits.target

print(f"Dataset shape: {X.shape}")
```

## 3. The Grand Comparison: PCA vs t-SNE vs UMAP

```python
fig, axes = plt.subplots(1, 3)

# 1. PCA
t0 = time.time()
pca = PCA(n_components=2).fit_transform(X)
axes[0].scatter(pca[:, 0], pca[:, 1], c=y, cmap='Spectral', s=10, alpha=0.8)
axes[0].set_title(f"PCA (Time: {time.time()-t0:.2f}s)")

# 2. t-SNE
t0 = time.time()
tsne = TSNE(n_components=2, perplexity=30, random_state=42).fit_transform(X)
axes[1].scatter(tsne[:, 0], tsne[:, 1], c=y, cmap='Spectral', s=10, alpha=0.8)
axes[1].set_title(f"t-SNE (Time: {time.time()-t0:.2f}s)")

# 3. UMAP
t0 = time.time()
umap_proj = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42).fit_transform(X)
scatter = axes[2].scatter(umap_proj[:, 0], umap_proj[:, 1], c=y, cmap='Spectral', s=10, alpha=0.8)
axes[2].set_title(f"UMAP (Time: {time.time()-t0:.2f}s)")

# Add colorbar
cbar = plt.colorbar(scatter, ax=axes, orientation='horizontal', fraction=0.02, pad=0.1)
cbar.set_label('Digit Label')

plt.suptitle("Dimensionality Reduction Comparison on Digits Dataset")
plt.show()
```

## 4. Hyperparameter Deep Dive: UMAP
Varying `n_neighbors`.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
neighbors = [5, 15, 50]

for i, n in enumerate(neighbors):
    proj = umap.UMAP(n_neighbors=n, min_dist=0.1, random_state=42).fit_transform(X)
    axes[i].scatter(proj[:, 0], proj[:, 1], c=y, cmap='Spectral', s=10, alpha=0.8)
    axes[i].set_title(f"UMAP (n_neighbors = {n})")

plt.show()
```
