# 📈 Time Series Forecasting Lab

Train an LSTM to forecast a multivariate time series.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate some fake multivariate stock data for demonstration
dates = pd.date_range("2020-01-01", periods=500)
data = {
    "Price": np.cumsum(np.random.randn(500)) + 100,
    "Volume": np.random.randint(1000, 5000, 500)
}
df = pd.DataFrame(data, index=dates)

df["Price"].plot(title="Synthetic Asset Price", figsize=(10,4))
plt.show()
```
