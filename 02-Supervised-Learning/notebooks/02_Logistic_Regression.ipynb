# 🚪 Logistic Regression

Despite its name, Logistic Regression is used for **classification** tasks (predicting discrete classes, like 0 or 1).

```
import numpy as np
from sklearn.linear_model import LogisticRegression

# Study hours vs Pass (1) or Fail (0)
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

log_reg = LogisticRegression()
log_reg.fit(X, y)

test_hours = np.array([[4.5]])
prediction = log_reg.predict(test_hours)
probability = log_reg.predict_proba(test_hours)

print(f"Prediction for 4.5 hours: {'Pass' if prediction[0] == 1 else 'Fail'}")
print(f"Probability: {probability[0][1]:.2f} (Pass)")
```
