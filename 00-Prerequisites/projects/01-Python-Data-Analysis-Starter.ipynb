# 📊 Python Data Analysis Starter

**Goal**: Load a dataset, perform basic exploratory data analysis (EDA), and visualize the results using Pandas and Matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)
```

### 1. Generating Synthetic Data
Instead of downloading a CSV, we will generate a synthetic dataset representing house prices to practice with.

```python
# Generate 500 rows of synthetic housing data
n_samples = 500

data = {
    'Square_Feet': np.random.normal(1500, 500, n_samples),
    'Bedrooms': np.random.randint(1, 6, n_samples),
    'Age_Years': np.random.randint(0, 50, n_samples),
}

# Price depends heavily on Square_Feet, slightly on Bedrooms, and decreases with Age
data['Price'] = (data['Square_Feet'] * 150) + (data['Bedrooms'] * 10000) - (data['Age_Years'] * 500) + np.random.normal(0, 10000, n_samples)

df = pd.DataFrame(data)

print("Data generated successfully!")
df.head()
```

### 2. Basic Data Exploration
Let's look at the shape of the data and some summary statistics.

```python
print(f"Dataset Shape: {df.shape}")
print("\nMissing Values:\n", df.isnull().sum())

# Get summary statistics
df.describe()
```

### 3. Data Visualization
Let's create some plots to understand the relationships between variables.

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot 1: Histogram of Prices
axes[0].hist(df['Price'], bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of House Prices')
axes[0].set_xlabel('Price ($)')
axes[0].set_ylabel('Count')
axes[0].grid(axis='y', alpha=0.75)

# Plot 2: Scatter plot of Square Feet vs Price
axes[1].scatter(df['Square_Feet'], df['Price'], alpha=0.5, color='coral')
axes[1].set_title('Price vs. Square Footage')
axes[1].set_xlabel('Square Feet')
axes[1].set_ylabel('Price ($)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### 4. Correlation Analysis
Which feature correlates the most with the Price?

```python
correlation_matrix = df.corr()

print("Correlation with Price:")
print(correlation_matrix['Price'].sort_values(ascending=False))
```
