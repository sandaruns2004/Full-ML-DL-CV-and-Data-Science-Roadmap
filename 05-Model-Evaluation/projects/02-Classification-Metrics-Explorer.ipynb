# Project 2: Classification Metrics Explorer\n\n**Objective**: Compare ROC vs PR curves on different class imbalances to see why PR curves are required for imbalanced data.

## Setup & Imports

```python
import numpy as np\nimport matplotlib.pyplot as plt\nfrom sklearn.datasets import make_classification\nfrom sklearn.linear_model import LogisticRegression\nfrom sklearn.model_selection import train_test_split\nfrom sklearn.metrics import roc_curve, precision_recall_curve, auc, average_precision_score
```

## Helper Function: Plotting Curves

```python
def plot_curves(weights, title):\n    X, y = make_classification(n_samples=5000, weights=weights, random_state=42)\n    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)\n    \n    model = LogisticRegression()\n    model.fit(X_train, y_train)\n    y_prob = model.predict_proba(X_test)[:, 1]\n    \n    fig, axes = plt.subplots(1, 2, figsize=(14, 5))\n    fig.suptitle(title, fontsize=16)\n    \n    # ROC\n    fpr, tpr, _ = roc_curve(y_test, y_prob)\n    roc_auc = auc(fpr, tpr)\n    axes[0].plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')\n    axes[0].plot([0, 1], [0, 1], 'k--')\n    axes[0].set_title('ROC Curve')\n    axes[0].set_xlabel('False Positive Rate')\n    axes[0].set_ylabel('True Positive Rate')\n    axes[0].legend()\n    \n    # PR\n    prec, rec, _ = precision_recall_curve(y_test, y_prob)\n    pr_auc = average_precision_score(y_test, y_prob)\n    baseline = sum(y_test) / len(y_test)\n    axes[1].plot(rec, prec, color='red', label=f'AP = {pr_auc:.2f}')\n    axes[1].axhline(y=baseline, color='k', linestyle='--')\n    axes[1].set_title('PR Curve')\n    axes[1].set_xlabel('Recall')\n    axes[1].set_ylabel('Precision')\n    axes[1].legend()\n    plt.show()
```

## Implementation & Comparison

```python
# 1. Balanced Dataset (50% / 50%)\nplot_curves([0.5, 0.5], "Scenario 1: Perfectly Balanced Data")\n\n# 2. Imbalanced Dataset (95% / 5%)\nplot_curves([0.95, 0.05], "Scenario 2: Highly Imbalanced Data (5% Minority)")
```

## Conclusion\nNotice how the ROC AUC barely drops (from 0.90 to 0.88) even though the model is fundamentally failing on the imbalanced dataset. The PR curve correctly identifies the failure, dropping from an AP of 0.91 to 0.35.
