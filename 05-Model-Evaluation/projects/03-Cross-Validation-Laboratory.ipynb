# Project 3: Cross-Validation Laboratory\n\n**Objective**: Prove the necessity of Stratified K-Fold mathematically by tracking the variance of test scores across folds on an imbalanced dataset.

## Setup & Imports

```python
import numpy as np\nfrom sklearn.model_selection import KFold, StratifiedKFold, cross_val_score\nfrom sklearn.datasets import make_classification\nfrom sklearn.ensemble import RandomForestClassifier\nimport matplotlib.pyplot as plt
```

## Simulating an Imbalanced Medical Dataset

```python
# 99% Healthy, 1% Sick\nX, y = make_classification(n_samples=500, weights=[0.99, 0.01], random_state=42)\nmodel = RandomForestClassifier(random_state=42)
```

## Standard K-Fold (Random)

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)\nkf_scores = cross_val_score(model, X, y, cv=kf, scoring='f1_macro')\nprint(f"Standard K-Fold Scores: {np.round(kf_scores, 3)}")\nprint(f"Standard K-Fold Variance: {np.var(kf_scores):.6f}")
```

## Stratified K-Fold (Proportional)

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)\nskf_scores = cross_val_score(model, X, y, cv=skf, scoring='f1_macro')\nprint(f"Stratified K-Fold Scores: {np.round(skf_scores, 3)}")\nprint(f"Stratified K-Fold Variance: {np.var(skf_scores):.6f}")
```

## Visualizing the Stability

```python
plt.figure(figsize=(8, 5))\nplt.plot(range(1, 6), kf_scores, marker='o', label='Standard K-Fold')\nplt.plot(range(1, 6), skf_scores, marker='s', label='Stratified K-Fold')\nplt.xticks(range(1, 6))\nplt.xlabel('Fold Number')\n\nplt.ylabel('Macro F1 Score')\nplt.title('Fold Stability Comparison')\nplt.legend()\nplt.grid(True)\nplt.show()
```

## Conclusion\nStandard K-Fold is highly unstable because some folds accidentally contain zero positive cases, crashing the F1 score. Stratified K-Fold mathematically prevents this, resulting in significantly lower variance.
