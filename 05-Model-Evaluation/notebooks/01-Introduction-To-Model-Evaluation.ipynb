# 01 - Introduction to Model Evaluation

Welcome to the first notebook in the Model Evaluation module. In this notebook, we will explore the fundamental concepts of why we evaluate models and demonstrate the illusion of 100% accuracy.

## Concept Overview
The biggest mistake a beginner can make is assuming that a model which performs perfectly on its training data will perform perfectly in the real world. This is called **Overfitting**, and it leads to **Data Leakage** illusions.

## Visual Demonstration
Let's generate a synthetic dataset and visualize it.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set_theme(style='whitegrid')

# Generate synthetic data
np.random.seed(42)
X = np.random.rand(200, 2) * 10
y = (X[:, 0] + X[:, 1] > 10).astype(int)

plt.figure(figsize=(8, 6))
plt.scatter(X[y==0][:, 0], X[y==0][:, 1], color='blue', label='Class 0 (Normal)', alpha=0.6)
plt.scatter(X[y==1][:, 0], X[y==1][:, 1], color='red', label='Class 1 (Anomaly)', alpha=0.6)
plt.title('Synthetic Classification Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

## Scikit-Learn Implementation
Now we will train a Decision Tree model on this exact data, and evaluate its accuracy *on the exact same data*.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Train the model
model = DecisionTreeClassifier(max_depth=None) # Unlimited depth allows perfect memorization
model.fit(X, y)

# Predict on the training data
predictions = model.predict(X)
accuracy = accuracy_score(y, predictions)

print(f'Training Accuracy: {accuracy * 100:.2f}%')
```

Wow! 100% Accuracy. We must be geniuses. 

But wait, what happens if we generate *new*, unseen data from the exact same mathematical distribution?

## Real Dataset Case Study (Simulation)

```python
# Generate new unseen data
X_new = np.random.rand(1000, 2) * 10
y_new = (X_new[:, 0] + X_new[:, 1] > 10).astype(int)

# Predict on unseen data
new_predictions = model.predict(X_new)
new_accuracy = accuracy_score(y_new, new_predictions)

print(f'Real World (Unseen) Accuracy: {new_accuracy * 100:.2f}%')
```

The accuracy drops because the tree learned the exact noise of the training set. It memorized the coordinates rather than learning the diagonal dividing line.

## Industry Notes
In production, if your model gets 99.9% accuracy on your first try, you almost certainly have Data Leakage. Always be paranoid of perfect scores.
