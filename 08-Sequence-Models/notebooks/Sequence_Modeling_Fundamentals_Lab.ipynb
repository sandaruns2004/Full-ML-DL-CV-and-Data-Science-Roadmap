# 🧬 Sequence Modeling Fundamentals Lab

In this notebook, we visualize the difference between independent data (tabular) and sequential data (time-series/text).

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="darkgrid")
np.random.seed(42)
```

### 1. Independent Data vs. Sequential Data

```
# Generate Independent Data (Standard Normal)
independent_data = np.random.randn(100)

# Generate Sequential Data (Random Walk / Autoregressive)
sequential_data = [0]
for i in range(1, 100):
    sequential_data.append(sequential_data[-1] + np.random.randn() * 0.5)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(independent_data, marker="o", linestyle="", alpha=0.7, color="blue")
axes[0].set_title("Independent Data (Order Does Not Matter)")

axes[1].plot(sequential_data, marker="o", linestyle="-", alpha=0.7, color="red")
axes[1].set_title("Sequential Data (Order & Time Matter)")

plt.tight_layout()
plt.show()
```

### 2. The Sliding Window Concept
To train models on sequences, we must chop continuous sequences into overlapping windows $(X)$ and targets $(y)$.

```
def create_windows(data, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)

window_size = 5
X, y = create_windows(sequential_data, window_size)

print("Original Sequence First 10:", np.round(sequential_data[:10], 2))
print("\nGenerated Windows:")
for i in range(3):
    print(f"Window {i+1}: X = {np.round(X[i], 2)} --> y = {np.round(y[i], 2)}")
```
