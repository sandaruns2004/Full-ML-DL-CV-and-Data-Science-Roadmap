# 🛍️ Customer Segmentation System

Using K-Means and Hierarchical Clustering to segment customers based on spending habits.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

np.random.seed(42)
income = np.random.normal(50000, 15000, 200)
spending_score = np.random.normal(50, 20, 200)
X = np.column_stack((income, spending_score))

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.title('Customer Segments')
plt.show()
```
