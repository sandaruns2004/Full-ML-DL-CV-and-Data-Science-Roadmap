# 🏘️ K-Nearest Neighbors Workshop
KNN is a simple algorithm: to classify a point, look at the K closest points to it and take a vote.

## 1. Problem Overview
We will classify wine types based on chemical analysis.

## 2. Dataset Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
display(df.head())
```

## 3. Data Preprocessing
Because KNN relies on distance (Euclidean), feature scaling is strictly required!

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## 4. From Scratch Implementation
Calculating Euclidean distance for a single test point against all training points.

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

test_point = X_test_scaled[0]
distances = [euclidean_distance(test_point, x_train) for x_train in X_train_scaled]
nearest_indices = np.argsort(distances)[:5] # k=5
votes = [y_train[i] for i in nearest_indices]
prediction = max(set(votes), key=votes.count)
print(f'Predicted: {prediction}, Actual: {y_test[0]}')
```

## 6. Scikit-Learn Workflow

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train_scaled, y_train)
print('Accuracy:', accuracy_score(y_test, model.predict(X_test_scaled)))
```

## 8. Hyperparameter Experiments
Testing different values of K.

```python
test_acc = []
ks = range(1, 25)
for k in ks:
    m = KNeighborsClassifier(n_neighbors=k)
    m.fit(X_train_scaled, y_train)
    test_acc.append(m.score(X_test_scaled, y_test))

plt.plot(ks, test_acc, marker='o')
plt.xlabel('K (Neighbors)')
plt.ylabel('Accuracy')
plt.title('Finding the optimal K')
plt.show()
```
