# 🔄 Transfer Learning: Fine-Tuning ResNet18
Adapting a pre-trained model to a new dataset.


```python
import torch
import torch.nn as nn
import torchvision
from torchvision import models, transforms
```

Download pre-trained ResNet18 and replace the final fully-connected layer:


```python
model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

# Freeze early layers
for param in model.parameters():
    param.requires_grad = False

# Replace classification head for 2 classes (e.g. Cats/Dogs or Ants/Bees)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 2)

print(model)
```
