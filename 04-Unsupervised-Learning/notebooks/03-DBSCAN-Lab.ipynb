Welcome to the DBSCAN Lab. We will compare K-Means and DBSCAN on complex, non-spherical datasets.

## 1. Theory Overview

DBSCAN groups together points that are closely packed together (points with many nearby neighbors), marking as outliers points that lie alone in low-density regions.

## 2. Visualization Playground
Let's see why K-Means fails on the Moons dataset and how DBSCAN solves it.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (14, 6)

X, _ = make_moons(n_samples=500, noise=0.05, random_state=42)
X_scaled = StandardScaler().fit_transform(X)

fig, axes = plt.subplots(1, 2)

# K-Means
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis', s=30)
axes[0].set_title("K-Means (Fails to capture geometry)")

# DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels, cmap='viridis', s=30)
axes[1].set_title("DBSCAN (Perfectly captures geometry)")

plt.show()
```

## 5. Scikit-Learn Implementation
DBSCAN automatically marks outliers as `-1`.

```python
# Adding some random noise to the data
noise = np.random.uniform(low=-2, high=2, size=(50, 2))
X_noisy = np.vstack([X_scaled, noise])

dbscan_noisy = DBSCAN(eps=0.3, min_samples=5)
labels_noisy = dbscan_noisy.fit_predict(X_noisy)

plt.figure(figsize=(8, 6))
plt.scatter(X_noisy[:, 0], X_noisy[:, 1], c=labels_noisy, cmap='plasma', s=30, edgecolor='k')
plt.title("DBSCAN handling noise (Purple dots are -1)")
plt.show()
```

## 6. Hyperparameter Impact Study
Varying Epsilon.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
eps_values = [0.1, 0.3, 0.8]

for i, eps in enumerate(eps_values):
    model = DBSCAN(eps=eps, min_samples=5)
    labels = model.fit_predict(X_noisy)
    axes[i].scatter(X_noisy[:, 0], X_noisy[:, 1], c=labels, cmap='tab20', s=30)
    axes[i].set_title(f"Epsilon = {eps} (Clusters: {len(set(labels)) - 1})")

plt.tight_layout()
plt.show()
```

## 7. Real Dataset Case Study
Use KD-Tree distances to find the optimal Epsilon using the K-distance graph.

```python
from sklearn.neighbors import NearestNeighbors

# Find 5th nearest neighbor distance for all points
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X_noisy)
distances, indices = neighbors_fit.kneighbors(X_noisy)

distances = np.sort(distances[:, 4], axis=0)

plt.figure(figsize=(8, 4))
plt.plot(distances)
plt.title("K-Distance Graph (Elbow indicates optimal Epsilon)")
plt.ylabel("5-NN Distance")
plt.xlabel("Data Points sorted by distance")
plt.axhline(y=0.3, color='r', linestyle='--')
plt.show()
```
