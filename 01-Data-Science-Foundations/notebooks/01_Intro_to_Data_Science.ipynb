Welcome to the first module of the Data Science Foundations!
Data Science is an interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from noisy, structured, and unstructured data.

## The Data Science Lifecycle
1. **Business Understanding**: Asking the right questions.
2. **Data Mining**: Scraping, APIs, SQL.
3. **Data Cleaning**: Handling missing values, outliers.
4. **Data Exploration (EDA)**: Visualizations, finding patterns.
5. **Feature Engineering**: Creating new variables from existing ones.
6. **Predictive Modeling**: Machine learning.
7. **Data Visualization**: Communicating results.

## A Quick Toy Example
Let's simulate some data and see how easily we can compute basic statistics in Python.

```python
# Generate 100 random ages between 18 and 65
np.random.seed(42)
ages = np.random.randint(18, 65, size=100)

print(f"First 10 ages: {ages[:10]}")
print(f"Average Age: {np.mean(ages):.2f}")
print(f"Oldest Person: {np.max(ages)}")
print(f"Youngest Person: {np.min(ages)}")
```
