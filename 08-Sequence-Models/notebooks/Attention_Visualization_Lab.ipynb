# 👁️ Attention Visualization Lab

In this notebook, we visualize how Attention weights matrix maps the relationships between inputs and outputs in a Sequence-to-Sequence model.

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Generating a Mock Attention Matrix
Imagine we are translating: "The black cat" (English) to "Le chat noir" (French). Notice how French inverts the order of the adjective and noun.

```
english_words = ["The", "black", "cat"]
french_words = ["Le", "chat", "noir"]

# Mock attention weights (rows = output french, cols = input english)
# Rows must sum to 1.0
attention_matrix = np.array([
    [0.90, 0.05, 0.05], # "Le" looks at "The"
    [0.05, 0.05, 0.90], # "chat" looks at "cat"
    [0.02, 0.96, 0.02], # "noir" looks at "black"
])

plt.figure(figsize=(6, 5))
sns.heatmap(attention_matrix, xticklabels=english_words, yticklabels=french_words, 
            cmap="Blues", annot=True, cbar_kws={"label": "Attention Weight"})
plt.title("Seq2Seq Attention Map")
plt.xlabel("Input Sequence (Encoder)")
plt.ylabel("Output Sequence (Decoder)")
plt.show()
```
