# 02 - Loss Landscapes & Gradient Descent
Visualize how gradient descent navigates the loss surface of a simple model.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Generate a convex loss surface (bowl shape)
W = np.linspace(-10, 10, 100)
b = np.linspace(-10, 10, 100)
W, b = np.meshgrid(W, b)
Loss = W**2 + b**2

plt.figure(figsize=(8, 6))
cp = plt.contourf(W, b, Loss, levels=20, cmap='viridis')
plt.colorbar(cp)
plt.title('Convex Loss Landscape (MSE)')
plt.xlabel('Weight (W)')
plt.ylabel('Bias (b)')
plt.plot([8, 5, 2, 0.5, 0], [8, 5, 2, 0.5, 0], marker='o', color='red', label='Gradient Descent Path')
plt.legend()
plt.show()
```
