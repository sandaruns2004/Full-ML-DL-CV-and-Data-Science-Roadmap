# 🖼️ Data Augmentation Visualizer
Seeing the effects of `torchvision.transforms` on images.


```python
import torch
import torchvision.transforms as T
from PIL import Image
import urllib.request
import matplotlib.pyplot as plt
```

```python
# Download a sample image
url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Cat03.jpg/1200px-Cat03.jpg'
urllib.request.urlretrieve(url, 'cat.jpg')
img = Image.open('cat.jpg').resize((256, 256))
```

```python
transforms = [
    ('Original', lambda x: x),
    ('Random Crop', T.RandomCrop(200)),
    ('Color Jitter', T.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0.5)),
    ('Random Rotation', T.RandomRotation(45)),
    ('Random Horizontal Flip', T.RandomHorizontalFlip(p=1.0))
]

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for ax, (name, transform) in zip(axes, transforms):
    ax.imshow(transform(img))
    ax.set_title(name)
    ax.axis('off')
plt.show()
```
